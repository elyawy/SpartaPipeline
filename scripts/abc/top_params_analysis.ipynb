```python
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use('default')
import seaborn as sns
sns.set(font_scale = 1.3)

sns.set_style("whitegrid")
sns.set_style("ticks")
```

```python
eggnog_top_params = "assets/all_top_params_eggnog.csv"
eggnog_df = pd.read_csv(eggnog_top_params)
yeast_top_params = "assets/all_top_params_yeast.csv"
yeast_df = pd.read_csv(yeast_top_params)
```

```python
eggnog_zipf_df = eggnog_df[eggnog_df.length_distribution == "zipf"]
eggnog_geometric_df = eggnog_df[eggnog_df.length_distribution == "geometric"]

yeast_zipf_df = yeast_df[yeast_df.length_distribution == "zipf"]
yeast_geometric_df = yeast_df[yeast_df.length_distribution == "geometric"]
```

```python
labels = list(eggnog_zipf_df.columns[:5])
```

```python
def plot_params_frequencies(data_df, file_name):
    fig = plt.figure(figsize=(15, 15))
    ax1 = plt.subplot(3,1,1)
    ax2 = plt.subplot(3,2,3)
    ax3 = plt.subplot(3,2,4)
    ax4 = plt.subplot(3,2,5)
    ax5 = plt.subplot(3,2,6)
    axes = [ax1, ax2, ax3, ax4, ax5]

    for idx, label in enumerate(labels):
        sns.histplot(data_df, x=label, bins=100, ax=axes[idx])
        axes[idx].set_xlabel("")
        axes[idx].set_title(" ".join(label.split("_")).title())
    fig.savefig(fname=file_name, bbox_inches="tight", dpi=300)
    
    # plt.show()
```

```python
plot_params_frequencies(yeast_zipf_df, "yeast_zipf_freqeuncies.png")
plot_params_frequencies(yeast_geometric_df, "yeast_geometric_freqeuncies.png")
plot_params_frequencies(eggnog_zipf_df, "eggnog_zipf_freqeuncies.png")
plot_params_frequencies(eggnog_geometric_df, "eggnog_geometric_freqeuncies.png")
```
